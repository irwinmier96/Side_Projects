```python
import pandas as pd
import datetime
```

```python
df = pd.read_csv("Datasets/us-counties.csv", encoding='utf-8')
df.head()
```

```python
df = df[['state', 'date', 'cases']]
df['date'] = pd.to_datetime(df.date)
df = pd.DataFrame(df)
df
```

```python
covid = df.groupby(['state','date'], as_index=False)['cases'].sum()
covid = covid.sort_values(['date'], ascending=False)
covid = pd.DataFrame(covid)
```

```python
#pivot df table and save it as us-states
us = pd.pivot_table(covid, values='cases', index=['state'], columns='date', fill_value=0)
us
```

```python
#Reset index and save to file for use in other notebooks
us = us.reset_index(level='state', col_level = 1)

```
