Applying the wordcloud function to all the articles in Canadian media to see how prevalent public health and coronavirus related topics are being covered. Built in stopwords are called in as well as other words that are adverbs and descriptive words that are too general are excluded from the wordcloud in order to see what are the most common nouns and adjectives that relate to the current meda coverage of the virus.

```python
#Wordcloud for all the text include a stopword 

#ca_text = " ".join(word for word in canada.text)

ca_text = ' '.join(canada['text'].str.lower())
print ("There are {} words in the combination of all review.".format(len(ca_text)))
```

```python

#Calling a set of built-in wtopwords and then added a new list of words to this list 
stop_words = set(stopwords.words('english'))  #stopwords.words is built in 
new_stopwords = ['said', 'said.', 'says', 'say', 'new', 'one','two', 'would',
                 'will', 'people', 'also', 'May', 'get', 'need', 'well', 'â\x80\x94'] #appending into stopwords.words
new_stopwords_list = stop_words.union(new_stopwords)

final_stop_words = set([word for word in new_stopwords_list if word in new_stopwords_list]) #set the final list before adding to the wordcloud

# stopwords.add(["one", "say", "time", "href", "https", "need", "people", "day", "may", "going", "Friday", "Tuesday",
#               "said", "per" , "cent", "says", "many", "still", "now", "thing", "Monday", "will", "part", "make",
#               "CBC", "New", "right", "viru", " country", "work", "take", "home", "even", "Thursday", "continue",
#               "want", "take", "city", "call", "March", "think", "well", "know", "way", "see", "go", "The", "-"])


wordcloud = WordCloud(stopwords = final_stop_words, background_color="white", max_words = 30).generate(ca_text)

# Display the generated image:
# the matplotlib way:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")

plt.savefig("Visualizations/Most_Common_Words_Canada_News.png", format="png")
plt.show()
```

Creating a wordcloud that takes the form of the Canadian flag. Much of the news is related to the provinces and the government of Canada, and mentions the current pandemic as well as health care related terms like confirmed cases and social distancing. In a graph highlighting the top ten words, most of the terms are realted to health care and the virus.

```python
#Make a wordcloud in the shape of the flag of Canada
mask = np.array(Image.open("Img/Canada.jpg"))
wordcloud_ca = WordCloud(stopwords=final_stop_words, background_color="white", mode="RGBA", max_words=1000, mask=mask).generate(ca_text)

# create coloring from image
image_colors = ImageColorGenerator(mask)
plt.figure(figsize=[12,12])
plt.imshow(wordcloud_ca.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")

plt.savefig("Visualizations/Most_Common_Words_Canada_News_Shape_Color_of_Canada_Flag.png", format="png")
plt.show()
```

```python
filtered_words = [word for word in ca_text.split() if word not in final_stop_words]
counted_words = collections.Counter(filtered_words)

words = []
counts = []
for letter, count in counted_words.most_common(10):
    words.append(letter)
    counts.append(count)
```

```python
words
```

```python
counts
```

```python
colors = cm.rainbow(np.linspace(0, 1, 10))
rcParams['figure.figsize'] = 20, 10

plt.title('Top Words that Appeared in Texts in Canadian News Sources During the Coronavirus Pandemic')
plt.xlabel('Count')
plt.ylabel('Words')
plt.barh(words, counts, color=colors)

plt.savefig("Visualizations/Top_Covid_Related_Words_in_Canadian_News")
```
