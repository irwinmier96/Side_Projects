```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Datasets/MoviesOnStreamingPlatforms_updated.csv")
```

```python
df = pd.DataFrame(df)
df = df.dropna( axis=0, how='any', thresh=None, subset=None, inplace=False)
df.head(5)
```

```python
df.dtypes
```

```python
#DF since 2010
years = [2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]

df = df[df["Year"].isin(years)]
df_2010s = pd.DataFrame(df)
df_2010s[10:20]

df_2010s.to_csv("Datasets/2010s_Titles")
```

```python
#Highest rated movies by age, make subplots
df_2010s['Title'] = df_2010s['Title'].astype('str')
df_2010s['Year'] = df_2010s['Year'].astype('str')
```

```python
df_2010s['IMDb'] = df_2010s['IMDb'].astype(float)
df_2010s['Rotten Tomatoes'] = df_2010s['Rotten Tomatoes'].str.rstrip('%').astype('float')
```

```python
title_year = df_2010s['Title'] + ', ' + df['Year']
df_2010s['Title (Year)'] = title_year
df_2010s = df_2010s[["Title (Year)", "Age", "IMDb", "Rotten Tomatoes", "Country"]]
df_2010s = pd.DataFrame(df_2010s)

df_2010s
```

```python
#Highest rated movies by age type
df_2010s_7 = pd.DataFrame(df_2010s[df_2010s["Age"] == "7+"])
df_2010s_all = pd.DataFrame(df_2010s[df_2010s["Age"] == "all"])
df_2010s_13 = pd.DataFrame(df_2010s[df_2010s["Age"] == "13+"])
df_2010s_16 = pd.DataFrame(df_2010s[df_2010s["Age"] == "16+"])
df_2010s_18 = pd.DataFrame(df_2010s[df_2010s["Age"] == "18+"])
```

```python
#Separate into IMDb and Rotten Tomatoes Rating
df_2010s_7_imbd = pd.DataFrame(df_2010s_7.sort_values(by="IMDb",ascending=False).head(10))
df_2010s_7_tomatoes = pd.DataFrame(df_2010s_7.sort_values(by="Rotten Tomatoes",ascending=False).head(10))

df_2010s_all_imbd = pd.DataFrame(df_2010s_all.sort_values(by="IMDb",ascending=False).head(10))
df_2010s_all_tomatoes = pd.DataFrame(df_2010s_all.sort_values(by="Rotten Tomatoes",ascending=False).head(10))

df_2010s_13_imbd = pd.DataFrame(df_2010s_13.sort_values(by="IMDb",ascending=False).head(10))
df_2010s_13_tomatoes = pd.DataFrame(df_2010s_13.sort_values(by="Rotten Tomatoes",ascending=False).head(10))

df_2010s_16_imbd = pd.DataFrame(df_2010s_16.sort_values(by="IMDb",ascending=False).head(10))
df_2010s_16_tomatoes = pd.DataFrame(df_2010s_16.sort_values(by="Rotten Tomatoes",ascending=False).head(10))

df_2010s_18_imbd = pd.DataFrame(df_2010s_18.sort_values(by="IMDb",ascending=False).head(10))
df_2010s_18_tomatoes = pd.DataFrame(df_2010s_18.sort_values(by="Rotten Tomatoes",ascending=False).head(10))
```

```python
from scipy.stats import linregress

lin_1 = df_2010s["IMDb"]
lin_2 = df_2010s["Rotten Tomatoes"]
```

```python
#Identify Averages 
avg_imdb = df_2010s["IMDb"].mean()
avg_tomatoes = df_2010s["Rotten Tomatoes"].mean()

print('Average IMDb Rating: ', avg_imdb  ,', ', 'Average Rotten Tomatoes Rating: ' , avg_tomatoes)

#Scatterplot Comparing IMDb and RT ratings
plt.figure(figsize=(10,5))
plt.scatter(df_2010s["IMDb"], df_2010s["Rotten Tomatoes"], marker = 'o', color = 'skyblue', edgecolor = 'black', alpha = 0.7)
plt.title("Average IMDb Rating vs Rotten Tomatoes Rating for Every Movie On Streaming Sites Since 2010", size = 20)

plt.xlabel("IMDb Rating", size = 15)
plt.ylabel("Rotten Tomatoes Rating", size = 15)

# Create gridlines
plt.axhline(y=62.9, color='grey', zorder=1, linewidth=0.75)
plt.axvline(x=6.29, color='grey', zorder=1, linewidth=0.75)
plt.grid(color = "grey",linestyle = 'dotted', linewidth = 0.5)

# Create a regression line
(slope, intercept, _, _, _) = linregress(lin_1, lin_2)
fit = slope * lin_1 + intercept
plt.plot(lin_1, fit, '--', zorder=3, color="green", linewidth=1)
plt.text(0, -40, "Note: Bolded grey lines mark the average IMDb and Rotten Tomatoes rating among all the movies since 2010 and the green dotted line is a linear regression line",
         fontsize=10)

plt.savefig("Visualizations/IMDb_Tomatoes_Scatter_and_Regression_Line_All_Movies_Since_2010")
plt.show()
```

```python
#Subplots
f = plt.figure(figsize=(20,20))
plt.title("Highest Rated Titles That Are Currently Streaming for Ages 7+", size=20)
plt.axis('off')

f.add_subplot(2,1,1)
plt.bar(df_2010s_7_imbd['Title (Year)'], df_2010s_7_imbd['IMDb'], alpha= 0.7, color = 'turquoise', edgecolor = 'black')
plt.ylabel("IMDb Rating", size = 15)
plt.xticks(rotation=20)

f.add_subplot(2,1, 2)
plt.bar(df_2010s_7_tomatoes['Title (Year)'], df_2010s_7_tomatoes['Rotten Tomatoes'], alpha= 0.7, color = 'violet', edgecolor = 'black')
plt.xlabel("Titles", size = 15)
plt.ylabel("Rotten Tomatoes", size = 15)
plt.xticks(rotation=20)

plt.savefig("Visualizations/Highest_Titles_For_Ages_7")
plt.show()
```
