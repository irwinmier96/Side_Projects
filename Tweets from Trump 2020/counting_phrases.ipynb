```python
import pandas as pd
import numpy as np
import re
import collections
import plotly.express as px
```

```python
df = pd.read_excel("Datasets/realDonaldTrumprecent2020.xlsx", sep='\t', 
                   names=["Tweet_ID", "Date", "Text"])
```

```python
df = pd.DataFrame(df)
df.head()
```

```python
tweets = df["Text"]

raw_string = ''.join(tweets)
no_links = re.sub(r'http\S+', '', raw_string)
no_unicode = re.sub(r"\\[a-z][a-z]?[0-9]+", '', no_links)
no_special_characters = re.sub('[^A-Za-z ]+', '', no_unicode)
no_capital_letters = re.sub('[A-Z]+', lambda m: m.group(0).lower(), no_special_characters)
```

```python
phrases = ['fake news', 'lamestream media', 'sleepy joe', 'radical left', 'bad ratings']
```

```python
#list of instances where the phrases show up
phrase_list = []
for phrase in phrases:
    for tweet in no_capital_letters:
        if tweet in phrase:
            phrase_list.append(tweet)
            
```
