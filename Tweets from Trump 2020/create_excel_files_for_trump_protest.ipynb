```python
import tweepy
import datetime
import xlsxwriter
import sys
import config
```

```python
#Connect to Twitter

##Possible ref: 
##https://blog.f-secure.com/how-to-get-tweets-from-a-twitter-account-using-python-and-tweepy/
consumerKey = config.api_key
consumerSecret = config.api_secret
accessToken = config.access_token
accessTokenSecret = config.access_token_secret
```

```python

auth = tweepy.OAuthHandler(consumerKey, consumerSecret)
auth.set_access_token(accessToken, accessTokenSecret)
```

```python
api = tweepy.API(auth)
```

```python
## Attribute Error issue:
## Solution: https://stackoverflow.com/questions/12906402/type-object-datetime-datetime-has-no-attribute-datetime
username = "realDonaldTrump"
startDate = datetime.datetime(2020, 5, 25, 0, 0, 0)
endDate =   datetime.datetime(2020, 6, 3, 12, 59, 59)
```

```python
tweets = []
tmpTweets = api.user_timeline(username, count = 40)
for tweet in tmpTweets:
    if tweet.created_at < endDate and tweet.created_at > startDate:
        tweets.append(tweet)
```

```python
tmpTweets
```

```python
while (tmpTweets[-1].created_at > startDate):
    tmpTweets = api.user_timeline(username, max_id = tmpTweets[-1].id)
    for tweet in tmpTweets:
        if tweet.created_at < endDate and tweet.created_at > startDate:
            tweets.append(tweet)
```

```python
workbook = xlsxwriter.Workbook(username + "sincemay.xlsx")
worksheet = workbook.add_worksheet()
row = 0
for tweet in tweets:
    worksheet.write_string(row, 0, str(tweet.id))
    worksheet.write_string(row, 1, str(tweet.created_at))
    worksheet.write(row, 2, tweet.text)
    row += 1

workbook.close()
print("Excel file ready")
```
