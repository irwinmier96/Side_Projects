```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
lgbt = pd.read_csv("Datasets/LGBT_Survey_DailyLife.csv")
lgbt = pd.DataFrame(lgbt)

lgbt.head()
```

```python
lgbt["question_label"].unique()
```

```python
lgbt['CountryCode'].unique()
```

```python
#create stacked bar charts by question and subset
b1_a_df = lgbt[lgbt['question_code'] == 'b1_a']
b1_a_df = pd.DataFrame(b1_a_df)

b1_a_df
```

```python
# del b1_a_df["question_label"]
# del b1_a_df["notes"]
b1_a_df
```

```python
#Compare answers in just Austria and then attempt other nations
b1_Austria_df = b1_a_df[b1_a_df["CountryCode"] == "Austria"]
b1_Austria_df = pd.DataFrame(b1_Austria_df)

b1_Austria_df["percentage"] = b1_Austria_df["percentage"].astype(str).astype(int)

b1_Austria_df
```

```python
fig, ax = plt.subplots(figsize=(25,18))  

answers = b1_Austria_df['answer'].drop_duplicates()
margin_bottom = np.zeros(len(b1_Austria_df['subset'].drop_duplicates()))
colors = ["orangered", "gold","aquamarine", "darkviolet", "palevioletred"]

for col, answer in enumerate(answers):
    values = list(b1_Austria_df[b1_Austria_df['answer'] == answer].loc[:, 'percentage'])

    b1_Austria_df[b1_Austria_df['answer'] == answer].plot.bar(x='subset',y='percentage', ax=ax, stacked=True, 
                                    bottom = margin_bottom, color=colors[col], label=answer)
    margin_bottom += values
    
    plt.grid()
    
    plt.xticks(rotation=45, fontsize = 20)
    
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
    
    plt.xlabel("Subsets for Austria", fontsize = 20)

    plt.ylabel("Percentage", fontsize = 20)
    
plt.title("Austria LGBT: In your opinion, how widespread is offensive language about lesbian, gay, bisexual and/or transgender people by politicians in the country where you live?", fontsize = 25)

plt.savefig("Visualizations/Austria_Question_b1_a_by_s")

plt.show()
```

```python
#Next barchart will focus on a certain question but will go by country with different subsets

lgbt_question = lgbt[lgbt['question_label'] == 'In your opinion, how widespread is public figures in politics, business, sports, etc being open about themselves being lesbian, gay, bisexual and/or transgender in the country where you live?']

lgbt_question = pd.DataFrame(lgbt_question)
lgbt_question.head()
```

```python
b1_g_gay = lgbt_question[lgbt_question["subset"] == 'Gay']
b1_g_lesbian = lgbt_question[lgbt_question["subset"] == 'Lesbian']
b1_g_bi_men = lgbt_question[lgbt_question["subset"] == 'Bisexual men']
b1_g_bi_women = lgbt_question[lgbt_question["subset"] == 'Bisexual women']
b1_g_trans = lgbt_question[lgbt_question["subset"] == 'Transgender']


b1_g_gay = pd.DataFrame(b1_g_gay)
b1_g_lesbian = pd.DataFrame(b1_g_lesbian)
b1_g_bi_men = pd.DataFrame(b1_g_bi_men)
b1_g_bi_women = pd.DataFrame(b1_g_bi_women)
b1_g_trans = pd.DataFrame(b1_g_trans)

## Bisexual men, bisexual women, and transgender people have a : instead of a number, replace that with 0
b1_g_bi_men.replace(to_replace=':', value=0, regex=True)
b1_g_bi_women.replace(to_replace=':', value=0, regex=True)
b1_g_trans.replace(to_replace=':', value=0, regex=True)

b1_g_gay.percentage=pd.to_numeric(b1_g_gay.percentage)
b1_g_lesbian.percentage=pd.to_numeric(b1_g_lesbian.percentage)
b1_g_bi_men.percentage=pd.to_numeric(b1_g_bi_men.percentage)
b1_g_bi_women.percentage=pd.to_numeric(b1_g_bi_women.percentage)
b1_g_trans.percentage=pd.to_numeric(b1_g_trans.percentage)

fig, ax = plt.subplots(figsize=(20,10))  

answers = b1_g_gay['answer'].drop_duplicates()
margin_bottom = np.zeros(len(b1_g_gay['CountryCode'].drop_duplicates()))
colors = ["orangered", "gold","aquamarine", "darkviolet", "palevioletred"]

for col, answer in enumerate(answers):
    values = list(b1_g_gay[b1_g_gay['answer'] == answer].loc[:, 'percentage'])

    b1_g_gay[b1_g_gay['answer'] == answer].plot.bar(x='CountryCode',y='percentage', ax=ax, stacked=True, 
                                    bottom = margin_bottom, color=colors[col], label=answer)
    margin_bottom += values
    
    plt.grid()
    
    plt.xticks(rotation=75, fontsize = 20)
    
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
    
    plt.xlabel("Countries in the EU", fontsize = 20)

    plt.ylabel("Percentage", fontsize = 20)
    
    plt.title("Gay: How widespread are public figures open about being LGBT in the country you live? ", fontsize = 25)

plt.savefig("Visualizations/Question_b1_g_Gay")

plt.show()
```
